## Barplot

Barplot is used to create bar charts, which are a type of graph that displays categorical data as rectangular bars with heights or lengths proportional to the values they represent. By looking at the relative heights or lengths of the bars, it might be possible to quickly compare the values and identify any trends or patterns. The basic barplot can be drawn as follows with matplotlib and plotly. But first, let me create a data for barplot examples:

```python
import random
from collections import Counter


random.seed(111)


x1 = [random.choice('ABCDE') for _ in range(20)]

let = Counter(x1)

print(let)
```

### Matplotlib

```python
import matplotlib.pyplot as plt

# preparing data for barplot
labels, values = zip(*let.items())

# drawing the graph
plt.bar(labels, values, color='skyblue')
plt.xlabel('Letters')
plt.ylabel('Frequency')
plt.title('Barplot')
plt.show()
```

```python
# Defining colors and legend texts
colors = ["red", "blue", "gray", "navy", "gold"]
legend_text = ["Group 1", "Group 2", "Group 3", "Group 4", "Group 5"]

# Creatinh a horizontal bar plot
plt.figure(figsize=(15, 8))
bars = plt.barh(labels, values, color=[colors[i % len(colors)] for i in range(len(values))], edgecolor='black')

# Adding title and labels
plt.title("Title of the plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")

# Setting x-axis limits
plt.xlim(0, 9)

# Adding a legend with distinct colors
# Creating a dummy bar plot for legend purposes
for color, label in zip(colors, legend_text):
    plt.plot([], [], color=color, label=label, linestyle='-', linewidth=10)  # Dummy line for legend

plt.legend(title="Groups", loc='upper right')

# Show the plot
plt.show()
```

## Pie Chart

A pie chart is a type of circular graph that is commonly used to represent data as a set of slices, where each slice corresponds to a category and its area or angle represents the proportion of that category relative to the whole. The total area or angle of the pie represents the total value of the data being displayed.

### Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Setting random seed for reproducibility
np.random.seed(111)

# Sample data
x1 = np.random.choice(list('ABCDE'), size=20, replace=True)
let, counts = np.unique(x1, return_counts=True)

# Calculating proportions
r1 = counts / len(x1)

# Creating labels with percentages
l1 = [f"Group-{name} {r:.1f}%" for name, r in zip(let, r1 * 100)]

# Defining colors
colors = ["red", "blue", "gray", "navy", "gold"]

# Creating a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=l1, colors=colors, autopct='%1.1f%%', startangle=140)

# Adding a title
plt.title("Title")

# Show the plot
plt.show()
```

### Plotly


```python
# Creating a pie chart
fig = go.Figure(data=[go.Pie(
    labels=l1,  # Labels for slices
    values=counts,  # Values for slices
    marker=dict(colors=colors),  # Colors for slices
    textinfo='label+percent',  # Display label and percentage
    rotation=140  # Rotate the pie chart
)])

# Adding a title
fig.update_layout(title="Title")

# Show the plot
fig.show()
```

Pie charts may not be the best choice for displaying complex data, and other visualization methods may be more appropriate.

They are limited to displaying proportions and do not show absolute values or counts.

## Histogram

A histogram is a graphical representation of the distribution of a dataset. It is a type of bar plot that represents the frequency or count of values that fall within a set of intervals or “bins”. The x-axis represents the intervals or bins of values, while the y-axis represents the frequency or count of values that fall within each interval. Bin width is an important detail that has been widely studied in the literature and is still debated. This is because deciding the number and width of the bins in a histogram can have a significant impact on the resulting visualization. Lets look at the significance of the bins with basic histograms. I will use mtcars dataset from seaborn library.

### Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the mtcars dataset from seaborn
df = sns.load_dataset('mpg').dropna()

# Extracting the 'mpg' column
mpg = df['mpg']

# Creating a histogram of the 'mpg' values
plt.figure(figsize=(10, 6))
plt.hist(mpg, bins=20, edgecolor='black', color='skyblue')

# Adding title and labels
plt.title('Histogram of MPG')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# if we change the bins:

# Create a histogram with 2 bins
plt.figure(figsize=(10, 6))
plt.hist(mpg, bins=2, edgecolor='black', color='skyblue')

# Add title and labels
plt.title('Histogram of MPG with 2 Bins')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Assuming that I wanted to analyse this histogram to have an idea about the distribution of the mpg variable, I would have different ideas about the distribution of the mpg variable for both graphs. For this reason, there are several methods for determining the appropriate number of bins, rules can be applied to hist function are:
- Sturges' formula: This method uses the following formula to determine the number of bins: k = 1 + log2(n), where k is the number of bins and n is the number of data points.
- Freedman-Diaconis rule: This method uses the interquartile range (IQR) to determine the width of each bin. The formula for the bin width is: bin width = 2 * IQR / (n^(1/3)), where n is the number of data points.
- Scott's rule: This method uses the standard deviation of the data to determine the width of each bin. The formula for the bin width is: bin width = 3.5 * sd / (n^(1/3)), where sd is the standard deviation of the data and n is the number of data points.

Here are some core arguments to draw fancier histograms:

- bins: Specifies the number of bins or their edges. Can be an integer or a sequence of bin edges.
- density: A boolean value indicating whether to normalize the histogram to form a probability density.
- color: Specifies the color of the bars. Can be a single color or a list of colors.
- edgecolor: Specifies the color of the edges of the bars.
- alpha: Controls the transparency of the bars. A value between 0 (transparent) and 1 (opaque).
- histtype: Defines the type of histogram. Options include 'bar', 'barstacked', 'step', and 'stepfilled'.
- range: Specifies the lower and upper range of the bins.
- align: Specifies how to align the bars. Options are 'left', 'mid', and 'right'.
- rwidth: Relative width of the bars as a fraction of the bin width.
- log: A boolean value indicating whether to use a logarithmic scale for the y-axis.


If I apply some of these changes to the previous histogram, it'll look like the following:

```python
# we first need to calculate Freedman-Diaconis rule for bin width. here is a function for that:
def freedman_diaconis_bins(data):
    # Calculating the interquartile range (IQR)
    q75, q25 = np.percentile(data, [75 ,25])
    iqr = q75 - q25
    # Calculating the bin width
    bin_width = 2 * iqr / (len(data) ** (1/3))
    # Calculating the number of bins
    num_bins = int((data.max() - data.min()) / bin_width)
    return num_bins

# Determining number of bins
num_bins = freedman_diaconis_bins(mpg)

# Plotting histogram
plt.figure(figsize=(12, 8))
plt.hist(mpg, bins=num_bins, edgecolor='yellow', color='black', density=False)

# Adding title and labels
plt.title('Histogram of MPG')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')

# Setting y-axis limit
plt.ylim(0, 0.25 * len(mpg))  # Adjusting the multiplier as needed based on actual data

# Hiding axes (in this context, this means not showing ticks and labels)
plt.gca().axes.get_yaxis().set_visible(False)  # it hides y-axis

# Showing the plot
plt.show()
```

Assume that you want to check a data for normality with a histogram, you can follow the following steps for matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(1212)

# Generating sample data
x = np.random.normal(loc=10, scale=4, size=1000)

# Sorting data
x = np.sort(x)

# Calculating density
mean_x = np.mean(x)
std_x = np.std(x)
fun = norm.pdf(x, loc=mean_x, scale=std_x)

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(x, bins=30, density=True, color='grey', edgecolor='black')

# Plotting density line
plt.plot(x, fun, color='black', linewidth=2)

# Adding title and labels
plt.title('Histogram of x')
plt.xlabel('x')
plt.ylabel('Density')

# Setting y-axis limit
plt.ylim(0, max(fun) * 1.1)

# Show the plot
plt.show()
```

## Boxplot

A boxplot is a graphical representation of the distribution of a dataset. It is a standardized way of displaying the distribution of data based on five summary statistics: the minimum value, the first quartile (Q1), the median, the third quartile (Q3), and the maximum value.

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['mpg'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            medianprops=dict(color='red'),
            flierprops=dict(markerfacecolor='red', marker='o'))

# Adding title and labels
plt.title('Boxplot of MPG')
plt.xlabel('Miles per Gallon (MPG)')

# Show the plot
plt.show()
```

Matplotlib’s boxplot() function offers a range of customization options. Here’s how you can use it with equivalent settings:


- color: Specifies the color of the boxes. Can be a single color or a list of colors for different boxes.
- boxprops: A dictionary of properties for the boxes. Can include color, linewidth, and other styling attributes.
- whiskerprops: A dictionary of properties for the whiskers. Includes color, linewidth, and other attributes.
- medianprops: A dictionary of properties for the median line. Includes color, linewidth, and other attributes.
- capprops: A dictionary of properties for the caps of the whiskers. Includes color, linewidth, and other attributes.
- flierprops: A dictionary of properties for the outliers. Includes color, marker, and other attributes.
- notch: A boolean value indicating whether to draw a notch around the median.
- vert: A boolean value indicating whether the boxes should be vertical (True) or horizontal (False).
- patch_artist: A boolean value indicating whether the boxes should be filled with color.
- widths: Width of the boxes. Can be a single value or a list of values.
- showfliers: A boolean value indicating whether to show outliers.


```python
# Creating a boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(df['mpg'], notch=True, vert=False, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'),
            medianprops=dict(color='red'),
            flierprops=dict(markerfacecolor='red', marker='o'),
            widths=0.5)

# Adding title and labels
plt.title('Boxplot of MPG')
plt.xlabel('Miles per Gallon (MPG)')

# Show the plot
plt.show()
```

- fillcolor: Specifies the fill color of the boxes. Can be a single color or a list of colors.

```python
# If we want to draw a plot that looks more fancy, we would draw something like the following:

# Loading iris dataset
df = sns.load_dataset('iris')

# Creating a boxplot grouped by species
fig = go.Figure()

for species, color in zip(df['species'].unique(), ['red', 'blue', 'gold']):
    fig.add_trace(go.Box(
        y=df[df['species'] == species]['sepal_width'],
        name=species,
        marker_color=color,
        boxmean='sd',  # standard deviation
        orientation='v'  # Vertical boxplot
    ))

# Adding a horizontal line at the mean of Sepal.Width
mean_sepal_width = df['sepal_width'].mean()
fig.add_shape(type='line',
              x0=-0.5, x1=2.5,
              y0=mean_sepal_width, y1=mean_sepal_width,
              line=dict(color='black', dash='dash'))

# Adding text annotation for the mean
fig.add_annotation(x=2, y=mean_sepal_width + 0.2, text=r'$\bar{x}$', showarrow=False, font=dict(size=14))

# Updating layout with title and labels
fig.update_layout(
    title='Sepal Widths by Species',
    xaxis_title='Species',
    yaxis_title='Sepal Width'
)

# Show the plot
fig.show()
```

## Line & Scatter Plots

### Matplotlib

```python
# Loading the mtcars dataset from seaborn
df = sns.load_dataset('mpg').dropna()

# Extracting the 'mpg' column
mpg = df['mpg']
# Plotting the 'mpg' values as a line plot
plt.plot(mpg)

# Adding title and labels
plt.title('Line Plot of MPG')
plt.xlabel('Index')
plt.ylabel('Miles per Gallon (MPG)')

# Show the plot
plt.show()
```

Advantages:

- Easy to create and read, making them an effective way to communicate data to a wide audience.

As a first plot, lets draw some chi squared datas with different degrees of freedom.

Matplotlib:

```python
from scipy.stats import chi2

# Defining the parameters
x = np.linspace(0, 60, 1000)
df = [5, 10, 20, 30]  # Degrees of freedom
colors = ["red", "blue", "green", "orange"]  # Line colors
line_types = ['--', '-', '-', '-']  # Line types (for linestyle)
line_widths = [1, 1, 1, 1]  # Line widths

# Creating an empty plot
plt.figure(figsize=(10, 6))
plt.xlim(0, 60)
plt.ylim(0, 0.)
plt.title("Different Chi-Square Distributions")
plt.xlabel("x")
plt.ylabel("Density")

# Plottting each Chi-Square distribution
for i in range(len(df)):
    fx = chi2.pdf(x, df[i])
    plt.plot(x, fx, color=colors[i], linestyle=line_types[i], linewidth=line_widths[i], label=f"df={df[i]}")

# Adding legend
plt.legend(title="Degrees of Freedom", loc="upper right")
plt.show()
```

```python
# plotly
x = np.linspace(0, 60, 1000)
df = [5, 10, 20, 30]  # Degrees of freedom
colors = ["red", "blue", "green", "orange"]  # Line colors
titles = [f"Chi-Square, df={d}" for d in df]  # Titles for each line
fig = go.Figure()

# Adding each Chi-Square distribution to the plot
for i in range(len(df)):
    fx = chi2.pdf(x, df[i])
    fig.add_trace(go.Scatter(
        x=x,
        y=fx,
        mode='lines',
        line=dict(color=colors[i], width=1, dash='solid'),  # Adjusting line width and color
        name=titles[i]
    ))

# Updating layout with titles and labels
fig.update_layout(
    title="Different Chi-Square Distributions",
    xaxis_title="x",
    yaxis_title="Density",
    xaxis_range=[0, 60],
    yaxis_range=[0, 0.2]
)

fig.show()
```

Now let me draw a more advaced scatter plot where each point is represented by a bubble whose size and color provide additional dimensions of information. This advanced visualization approach allows for a more nuanced understanding of the data, showing not just relationships between two variables but also adding context with size and color. The use of transparency and edge lines ensures that the plot remains clear and interpretable even when data points are densely packed.

```python
# Generating sample data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.abs(np.random.randn(100) * 100)  # Bubble sizes
colors = np.random.rand(100)  # Bubble colors
alpha = 0.7  # Transparency
cmap = plt.get_cmap('viridis')  # Color map

# Creating the plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    x, y,
    s=sizes,  # Bubble sizes
    c=colors,  # Bubble colors
    alpha=alpha,  # Transparency
    cmap=cmap,  # Color map
    edgecolor='k',  # Edge color of bubbles
    linewidth=0.5  # Edge line width
)

# Adding titles and labels
plt.title('Advanced Scatter Plot', fontsize=16, fontweight='bold')
plt.xlabel('X Axis Label', fontsize=14)
plt.ylabel('Y Axis Label', fontsize=14)

# Adding a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Color Scale', fontsize=14)
cbar.ax.tick_params(labelsize=12)

# Adding grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Adding a legend
plt.legend(['Data Points'], loc='best', fontsize=12)

plt.tight_layout()
plt.show()
```
